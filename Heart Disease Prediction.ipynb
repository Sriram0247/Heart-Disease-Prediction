**Importing relevant libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.neighbors import KNeighborsClassifier
```

**Import dataset**

```python
df = pd.read_csv('heart.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

**Feature selection**

```python
corrmat = df.corr()
top_corr_features = corrmat.index
```

**Plot heat map**

```python
plt.figure(figsize = (16,16))
g = sns.heatmap(df[top_corr_features].corr(), annot = True, cmap = "RdYlGn")
plt.show()
```

Next, we compare the correlation among features and remove one of the two features that have correlation higher than 0.9

```python
columns = np.full((corrmat.shape[0],), True, dtype = bool)
for i in range(corrmat.shape[0]):
    for j in range(i+1, corrmat.shape[0]):
        if corrmat.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
                
selected_columns = df.columns[columns]

df = df[selected_columns]
```

```python
df.info()
```

We observe that none of the columns have been removed. Hence, we retain all the features.

Finally, we want to ensure whether the target classes are of approximately equal size.

```python
sns.set_style('whitegrid')
sns.countplot(x = 'target', data = df, palette ='RdBu_r')
plt.show()
```

**Data processing**

**One Hot Encoding / Dummy Encoding of categorical variables**

```python
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
dataset.head()
```

**Scaling the numerical variables**

We observe that the numerical variables vary in their scales. Hence, we standardize the data before proceeding forward.

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
dataset.head()
```

**Creating a feature set (X) and outcome variable (Y)**

```python
Y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

**Performing KNN classification and finding the optimum n_neighbors**

With the feature set and outcome variables created, we finally perform the k-nearest neighbors classification. To find the optimum number of k-nearest neighbors, we iterate the process of fitting the KNN classifier for values of k from 1 to 20. We fit the model using 10-fold cross-validation, calculate the accuracy of the model ('score') for each fold and then calculate the average score of all folds. This average score is then recorded in a list ('knn_scores') as the score for a particular value of 'k'.

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier,X,Y,cv = 10)
    knn_scores.append(score.mean())
```

For values of 'k' from 1 to 20, we plot the average scores from the list 'knn_scores'.

**Plot the KNN classifier scores for different values of n_neighbors**

```python
plt.figure(figsize = (16,16))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1, 21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
plt.show()
```

We observe that model accuracy is highest for k = 9. We can see that the model achieved an estimated classification accuracy of about 84.49%.

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 9)
score = cross_val_score(knn_classifier, X, Y, cv = 10)
score.mean()
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators=10)
score = cross_val_score(randomforest_classifier, X, Y, cv=10)
score.mean()
```

A similar exercise performed with the random forest classifier gives a model accuracy of 81.12%. Hence, we conclude that the KNN classifier is a better predictor than the random forest classifier.
